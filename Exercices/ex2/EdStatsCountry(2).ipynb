```python

# Étape 3 – Nettoyage des faux pays

# Étude des valeurs uniques dans la colonne "Region"
print("Valeurs uniques dans 'Region' :")
print(EdStatsCountry['Region'].unique())

# Faux pays : ceux sans région ou avec des regroupements statistiques
faux_pays_df = EdStatsCountry[
    EdStatsCountry['Region'].isnull() |
    EdStatsCountry['Region'].str.contains("income|World|OECD", case=False, na=False)
]

# Liste des codes des faux pays
faux_pays_codes = faux_pays_df['Country Code'].tolist()
print("Nombre de faux pays détectés :", len(faux_pays_codes))

# Nettoyage de la table Country
EdStatsCountry_clean = EdStatsCountry[~EdStatsCountry['Country Code'].isin(faux_pays_codes)].copy()
print("Pays restants après nettoyage :", EdStatsCountry_clean.shape[0])
```

```python
print(EdStatsData.columns)
```

```python
EdStatsData_clean = EdStatsData[~EdStatsData['Country Code'].isin(faux_pays_codes)].copy()
EdStatsCountry_Series_clean = EdStatsCountry_Series[~EdStatsCountry_Series['CountryCode'].isin(faux_pays_codes)].copy()
EdStatsFootNote_clean = EdStatsFootNote[~EdStatsFootNote['CountryCode'].isin(faux_pays_codes)].copy()
```

```python
# On ne garde que les pays valides en faisant une jointure avec EdStatsCountry_clean

# Jointure sur EdStatsData (clé = 'Country Code')
EdStatsData_clean_merge = EdStatsData.merge(
    EdStatsCountry_clean[['Country Code']],
    on='Country Code',
    how='inner'
)

# Jointure sur EdStatsCountry_Series (clé = 'CountryCode' ≠ nom dans EdStatsCountry)
EdStatsCountry_Series_clean_merge = EdStatsCountry_Series.merge(
    EdStatsCountry_clean[['Country Code']],
    left_on='CountryCode',
    right_on='Country Code',
    how='inner'
)

# Jointure sur EdStatsFootNote (clé = 'CountryCode')
EdStatsFootNote_clean_merge = EdStatsFootNote.merge(
    EdStatsCountry_clean[['Country Code']],
    left_on='CountryCode',
    right_on='Country Code',
    how='inner'
)
```

```python
print("Méthode liste - Data :", EdStatsData_clean.shape)
print("Méthode merge - Data :", EdStatsData_clean_merge.shape)

print("Méthode liste - Series :", EdStatsCountry_Series_clean.shape)
print("Méthode merge - Series :", EdStatsCountry_Series_clean_merge.shape)

print("Méthode liste - FootNote :", EdStatsFootNote_clean.shape)
print("Méthode merge - FootNote :", EdStatsFootNote_clean_merge.shape)
```
